```python
import pandas as pd 
import numpy as np
```

```python
#set display options
pd.set_option('display.max_columns', None)
```

```python
#load in csv's for whole year individually. Using JC Data due to Tableau Size limit

June = pd.read_csv("Data/JC Data/JC-202006-citibike-tripdata.csv")
July = pd.read_csv("Data/JC Data/JC-202007-citibike-tripdata.csv")
August = pd.read_csv("Data/JC Data/JC-202008-citibike-tripdata.csv")
September = pd.read_csv("Data/JC Data/JC-202009-citibike-tripdata.csv")
October = pd.read_csv("Data/JC Data/JC-202010-citibike-tripdata.csv")
November = pd.read_csv("Data/JC Data/JC-202011-citibike-tripdata.csv")
December = pd.read_csv("Data/JC Data/JC-202012-citibike-tripdata.csv")
January = pd.read_csv("Data/JC Data/JC-202101-citibike-tripdata.csv")
February = pd.read_csv("Data/JC Data/JC-202102-citibike-tripdata.csv")
March = pd.read_csv("Data/JC Data/JC-202103-citibike-tripdata.csv")
April  = pd.read_csv("Data/JC Data/JC-202104-citibike-tripdata.csv")
May = pd.read_csv("Data/JC Data/JC-202105-citibike-tripdata.csv")

```

```python
#Cleaning Data! fist up date 

# Feb 2021 starts using simple time cut off after decimal
# need to make all the same format
# 2020-06-01 01:19:59.2090 chop off/convert to 2021-02-03 23:11:28

months_that_need_time_chop = [June, July, August, September, October, November, December, January]

columns_that_need_time_chop = ['starttime', 'stoptime']

# It's Processing Time!

# take off the last 5 characters columns_that_need_time_chop
for month in months_that_need_time_chop:
    for column in columns_that_need_time_chop:
        month[column] = month[column].str[:-5] 
```

```python
# Combine months into one big dataframe!

dataframes = [June, July, August, September, October, November, December, January, February, March, April, May]
combined = pd.concat(dataframes)
combined
```

```python
# Combine columns to get rid of duplicates due to naming conventions. 

combined["started_at"] = combined["started_at"].combine_first(combined['starttime'])
combined['ended_at'] = combined['ended_at'].combine_first(combined["stoptime"])
combined['start_station_name'] = combined['start_station_name'].combine_first(combined['start station name'])
combined['end_station_name'] = combined['end_station_name'].combine_first(combined['end station name'])
combined['start_lat'] = combined['start_lat'].combine_first(combined['start station latitude'])
combined['end_lat'] = combined['end_lat'].combine_first(combined['end station latitude'])
combined['start_lng'] = combined['start_lng'].combine_first(combined['start station longitude'])
combined['end_lng'] = combined['end_lng'].combine_first(combined['end station longitude'])
combined['member_casual'] = combined['member_casual'].combine_first(combined['usertype'])
combined 

# drop things we don't need 
cleaned = combined.drop(labels=['start station name', 'end station name', 'usertype', 'birth year', 
                                "tripduration", 'starttime', 'stoptime', 'start station id', 
                                'start station latitude', 'start station longitude', 'rideable_type', 
                                'end station id', 'end station latitude', 'end station longitude', 
                                'bikeid', 'gender', 'ride_id', 'start_station_id', 'end_station_id' ], axis = 1)
cleaned
```

```python
# Export Dataframe to CSV 

cleaned.to_csv("Data/cleaned_JC.csv")
```
